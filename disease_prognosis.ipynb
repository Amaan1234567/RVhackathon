```python
from sklearn.model_selection import train_test_split
df=pd.read_csv('Training.csv')
""" test_df=pd.read_csv('Testing.csv')
fr=[df,test_df]

final_df=pd.concat(fr,join='outer')
final_df.drop(columns=['Unnamed: 133'],inplace=True)
final_df """
df.drop(columns=['Unnamed: 133'],inplace=True)
df['prognosis_encode']=df['prognosis'].apply(encode)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['prognosis','prognosis_encode']),df['prognosis_encode'], test_size=0.3,random_state=101)
print(list(X_test.iloc[0]))
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,pred),'\n',classification_report(y_test,pred))
```
